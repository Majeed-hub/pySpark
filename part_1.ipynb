```python
import pyspark
```

```python
from pyspark.sql import SparkSession
```

```python
spark=SparkSession.builder.appName("Practise").getOrCreate()
```

```python
df_spark=spark.read.csv("student.csv")
df_spark.show()
```

```python
df_spark=spark.read.option("header","true").csv("student.csv")
df_spark.show()
```

```python
## To check the schema
df_spark.printSchema()
```

```python
## To correct schema which takes all values as strings
df_spark=spark.read.option("header","true").csv("student.csv",inferSchema=True)
df_spark.printSchema()
```

```python
## We can simplify the reading query
df_spark=spark.read.csv("student.csv",inferSchema=True,header=True)
df_spark.show()
```

```python
## to see columns 
df_spark.columns
```

```python
## To see values of a particular column
df_spark.select("name").show()
```

```python
## To select multiple columns at once 
df_spark.select(["name","class"]).show()
```

```python
df_spark.dtypes
```

```python
df_spark.describe().show()
```

```python
## adding colomn in a dataframe
df_spark.withColumn("age",df_spark["rollno"]*5).show()
```

```python
## Dropping column in a data frame
df_spark.drop("age").show()
```

```python
### Renaming columns 
df_spark.withColumnRenamed("class","Degree").show()
```
